Στην Ευρώπη οι τελευταίες καταγραφές αφορούν την περίοδο 2011 έως 2019, με τιμές που διαφέρουν ανά χώρα, εξετάζοντας το φύλο και την ηλικία.

Οι πίνακες αφορούν με την σειρά τις χώρες : Σουηδία, Πορτογαλία, Ελλάδα , Τσεχία. Η μπλε γραμμή απεικονίζει την πορεία θανατών για τους άνδρες και η πορτοκαλί γραμμή για τις γυναίκες. Ενώ και στις 4 χώρες τα ποσοστά αυτοκτονιών των γυναικών σημειώνουν παρόμοια πορεία , στους άνδρες υπάρχουν ευδιάκριτες διαφορές. 

Στην Σουήδια παρατηρούνται έντονες αυξομειώσεις κυρίως στα ποσοστά των ανδρών. Και στα δύο φύλα σημειώνεται αύξηση κατά το έτος 2013, ενώ το έτος 2015 παρατηρούμε αντιθετική εξέλιξη, με τους άνδρες να παρουσιάζουν αύξηση και τις γυναίκες μείωση. Το 2014 ήταν η χρονιά που στην Πορτογαλία παρατηρήθηκε μεγαλύτερο ποσοστό αυτοκτονιών ως προς τους συνολικούς θανάτους, συγκριτικά με τα άλλα έτη.Η Ελλάδα φαίνεται να είναι η χώρα με τα χαμηλοτέρα ποσοστά, ενώ η Τσεχία η χώρα με τα υψηλότερα. Παρόλα αυτά , και στις τέσσερεις χώρες όσον αφορά τους άνδρες τα ποσοστά πλησιάζουν το 30%, ενώ στις γυναίκες δεν ξεπερνούν το 10%.

"13 reasons why" -2017 σειρά του Netflix- εξιστορεί την ιστορία μιας έφηβης που δίνει τέλος στη ζωή της και αφήνει πίσω 13 κασέτες που ηχογράφησε πριν αυτοκτονήσει. "Μπλε φάλαινα": ένα παιχνίδι από σειρά καθηκόντων που ανατίθονται σε εφήβους, με τελική πρόκληση την διάπραξη αυτοκτονίας. Η συχνότητα τέτοιων παραδειγμάτων στις ζωές των εφήβων, ίσως ήταν ένας λόγος που θα μπορούσε να δικαιολογήσει τα ποσοστά αυτοκτονιών στις ηλικίες 15-19. 

Στον παρακάνω χάρτη η τονικότητα των χρωμάτων δηλώνει τα ποσοστά αυτοκτονιών ως προς το συνολικό αιρθμό θανάτων στην συγκεκριμένη ηλικιακή κλίμακα. Όσο πιο έντονο το χρώμα κάθε χώρας, τόσο μεγαλύτερο το ποσοστό.

Στις τέσσερεις επιλεγμένες χώρες τα ποσοστά της ηλικιακής κλίμακας 15-19 , κυμαίνονται πιο χάμηλα συγκριτικά με τα συνολικά ποσοστά ανδρών και γυναικών. 


Στους παρακάτω πίνακες παρατηρούμε ότι η Ελλάδα και η Πορτογαλία σημειώνουν τα χαμηλότερα ποσοστά, ενώ η Σουηδία και η Τσεχία τα υψηλότερα. 


Η αυτοκτονία αποτελεί ένα πολυπαραγοντικό φαινόμενο που δεν μπορεί να αποδοθεί σε μονοδιάστατα αίτια. Ένας ευρέως διαδεδομένος μύθος είναι ότι «αυτοί που εκφράζουν την πρόθεση τους να πεθάνουν δεν πρόκειται να το κάνουν». Ωστόσο, πρέπει να παίρνουμε πάντα στα σοβαρά τα αυτοκαταστροφικά σχόλια. "Το 95% των αυτοκτονιών θα μπορούσε να προληφθεί μέσω της γνώσης και της αναγνώρισης των προειδοποιητικών σημείων και συμπτωμάτων", αναφέρουν η ψυχολόγος Ντόροθυ Γαρώνη και η κλινική ψυχολόγος Τζίνα Κεφάλα από το Κέντρο Ημέρας για την πρόληψη της Αυτοκτονίας της «Κλίμακα», στη LIFO και τον Ευθύμιο Σαββάκη. Το μεγαλύτερο όπλο πρόληψης, αν όχι για την πλήρη εξάλειψη αλλά τουλάχιστον για την μείωση του φαινομένου, είναι αδιαμφισβήτητα η ψυχοθεραπεία και γενικοτέρα η απευθύνση σε ειδικούς.

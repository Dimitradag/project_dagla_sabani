"Μια έρευνα για τα ποσοστά αυτοκτονιών στην Ευρώπη"

Πληκτρολογώντας στην μηχανή αναζήτησης της Google την λέξη "αυτοκτονίες", μείναμε έκπληκτες με τις πιο συχνές αναζητήσεις που εμφάνιζε.

"Αλεξάντερ Μακ Κουίν: Βυθισμένος στην κατάθλιψη ο διάσημος σχεδιαστής μόδας είχε προσπαθήσει να πάρει τη ζωή του με υπερβολική δόση ναρκωτικών ήδη δύο φορές πριν την τρίτη και επιτυχημένη του προσπάθεια."

"Μέριλιν Μονρόε: βρέθηκε νεκρή -και γυμνή- στο κρεβάτι της στις 4 Αυγούστου 1962, από υπερβολική δόση βαρβιτουρικών."

Πρόσφατη είδηση: «Έχω κλειδώσει τις πόρτες και θα αυτοκτονήσω» Αυτές φαίνεται να ήταν οι τελευταίες λέξεις του 16χρονου που έβαλε τέλος στη ζωή του μέσα στο μπάνιο του σπιτιού του, στο Παλαιό Φάληρο.

Οι ειδήσεις αυτοκτονιών ανά τον κόσμο είναι ένα φαινόμενο που εξελίσσεται μέσα στα χρόνια. Άλλοτε με υψηλότερα ποσοστά, άλλοτε με χαμηλότερα αλλά πάντα υπαρκτό. Στην Ευρώπη οι τελευταίες καταγραφές αφορούν την περίοδο 2011 έως 2019, με τιμές που διαφέρουν ανά χώρα, εξετάζοντας το φύλο και την ηλικία.

Στην έρευνα μας , αναλύοντας δεδομένα από όλες τις χώρες της Ευρώπης, καταλήξαμε στο ότι η χώρα με τα μεγαλύτερα ποσοστά αυτοκτονιών είναι το Ηνωμένο Βασίλειο κατά το έτος 2019 , με ποσοστό 66.31% στον ανδρικό πληθυσμό."Οι άνδρες έχουν την τάση να πιστεύουν ότι το να εκφράζουν τα συναισθήματα τους είναι μία ένδειξη αδυναμίας. [...] Αυτό συχνά ξεκινάει από την παιδική ηλικία με την γνωστή φράση -οι άνδρες δεν κλαίνε-.", αναφέρει ο Nick Gibbens, εκρόσωπος για την ευαισθητοποίηση της ψυχικής υγείας των ανδρών και την πρόληψη των αυτοκτονιών. Παρόμοια και στις ηλικίες 15-19 ετών , το Ηνωμένο Βασίλειο κατέχει και πάλι την πρώτη θέση με ποσοστό 22.79%. 

Προκειμένου να έχουμε πιο έγκυρα αποτελέσματα, στην έρευνα που εκπονήσαμε αναλύσαμε πως διαφέρουν τα ποσοστά αυτοκτονιών ανά φύλο σε 4 ευρωπαικές χώρες με παρόμοιο πληθυσμό.




Οι πίνακες αφορούν με την σειρά τις χώρες : Σουηδία, Πορτογαλία, Ελλάδα , Τσεχία. Η μπλε γραμμή απεικονίζει την πορεία θανατών για τους άνδρες και η πορτοκαλί γραμμή για τις γυναίκες. Ενώ και στις 4 χώρες τα ποσοστά αυτοκτονιών των γυναικών σημειώνουν παρόμοια πορεία , στους άνδρες υπάρχουν ευδιάκριτες διαφορές. 

Στην Σουήδια παρατηρούνται έντονες αυξομειώσεις κυρίως στα ποσοστά των ανδρών. Και στα δύο φύλα σημειώνεται αύξηση κατά το έτος 2013, ενώ το έτος 2015 παρατηρούμε αντιθετική εξέλιξη, με τους άνδρες να παρουσιάζουν αύξηση και τις γυναίκες μείωση. Το 2014 ήταν η χρονιά που στην Πορτογαλία μεγάλο μέρος του πληθυσμού που αφόρα και τα δύο φύλα, κατέφυγε στην αυτοκτονία. Η Ελλάδα φαίνεται να είναι η χώρα με τα χαμηλοτέρα ποσοστά, ενώ η Τσεχία η χώρα με τα υψηλότερα. Παρόλα αυτά , και στις τέσσερεις χώρες όσον αφορά τους άνδρες τα ποσοστά αγγίζουν εως και το 20%, ενώ στις γυναίκες δεν ξεπερνούν το 10%. 

"13 reasons why" -2017 σειρά του Netflix- εξιστορεί την ιστορία μιας έφηβης που δίνει τέλος στη ζωή της και αφήνει πίσω 13 κασέτες που ηχογράφησε πριν αυτοκτονήσει. "Μπλε φάλαινα": ένα παιχνίδι από σειρά καθηκόντων που ανατίθονται σε εφήβους, με τελική πρόκληση την διάπραξη αυτοκτονίας. Η δημοφιλία τέτοιων παραδειγμάτων που αφορούν εφήβους, ίσως οφείλεται για τα ποσοστά αυτοκτονιών στις ηλικίες 15-19.

Στον παρακάνω χάρτη η τονικότητα των χρωμάτων δηλώνει τα ποσοστά. Όσο πιο έντονο το χρώμα κάθε χώρας, τόσο μεγαλύτερο το ποσοστό.

Στις τέσσερεις επιλεγμένες χώρες τα ποσοστά της ηλικιακής κλίμακας 15-19 , κυμαίνονται ανάλογα όπως στον υπόλοιπο πλυθησμό. Με την Ελλάδα και την Πορτογαλία να σημειώνουν τα χαμηλότερα ποσοστά, ενώ η Σουηδία και η Τσεχία τα υψηλότερα.  

Αντίστοιχα τα ραβδογράμματα αφορούν με την σειρά τις χώρες: Σουηδία, Πορτογαλία, Ελλάδα, Τσεχία.


Η αυτοκτονία αποτελεί ένα πολυπαραγοντικό φαινόμενο που δεν μπορεί να αποδοθεί σε μονοδιάστατα αίτια. Ένας ευρέως διαδεδομένος μύθος είναι ότι «αυτοί που εκφράζουν την πρόθεση τους να πεθάνουν δεν πρόκειται να το κάνουν». Ωστόσο, πρέπει να παίρνουμε πάντα στα σοβαρά τα αυτοκαταστροφικά σχόλια. "Το 95% των αυτοκτονιών θα μπορούσε να προληφθεί μέσω της γνώσης και της αναγνώρισης των προειδοποιητικών σημείων και συμπτωμάτων", αναφέρουν η ψυχολόγος Ντόροθυ Γαρώνη και η κλινική ψυχολόγος Τζίνα Κεφάλα από το Κέντρο Ημέρας για την πρόληψη της Αυτοκτονίας της «Κλίμακα», στη LIFO και τον Ευθύμιο Σαββάκη.

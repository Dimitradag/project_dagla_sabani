```python
#import pandas and call it pd from this moment onwards in this notebook
import pandas as pd 
#import matplotlib and call it plt from this moment onwards in this notebook
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#download sex_table as excel in our pc
sex_table.to_csv('sex_table.csv',index=False)
```

```python
#read your csv
df = pd.read_csv('sex_table.csv')

#create a new df, called df1
#this will be your initial df, grouped by a list of things: COUNTRY, TIME_PERIOD, SEX, calculating the median PERCENTAGE. This is just a trick: you are not calculating any median, since there is only one percentage per sex in a country in a given time period. You are just using median() in order to get the real value of the percentage in each grouped case. 
df1 = pd.DataFrame(df.groupby(['COUNTRY','TIME_PERIOD','SEX']).PERCENTAGE.median()).reset_index()

```

```python
df1
```

```python
#create plot for one country and one sex
ax = df1[(df1.COUNTRY=='EL')&(df1.SEX=='M')].plot(kind='line',x='TIME_PERIOD',y='PERCENTAGE',color='blue')
#then, overlay with another plot for the other sex and the same country
df1[(df1.COUNTRY=='EL')&(df1.SEX=='F')].plot(kind='line',x='TIME_PERIOD',y='PERCENTAGE',color='orange',ax=ax)
```

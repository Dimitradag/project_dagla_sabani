```python
#import pandas and call it pd from this moment onwards in this notebook
import pandas as pd
```

```python
#Read csv using pandas
df=pd.read_csv("tps00202_page_linear.csv")
```

```python
#df with all columns label and all rows of csv
df.columns=df.iloc[0]
```

```python
#print out this dataframe
df
```

```python
#print out this dataframe
df
```

```python
#print out this dataframe
df1
```

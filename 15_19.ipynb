```python
#import pandas and call it pd from this moment onwards in this notebook
import pandas as pd
```

```python
#Read csv using pandas
df=pd.read_csv("tps00202_page_linear.csv")
```

```python
#df with all columns label and all rows of csv
df.columns=df.iloc[0]
```

```python
#print out this dataframe
df
```

```python
#dropping passed columns
df.drop(["ESTAT:TPS00202(1.0)", "09/12/22 23:00:00", "A", "RT", "X60-X84_Y870"], axis = 1, inplace = True)
```

```python
#print out this dataframe
df
```

```python
# delete the missing values and the last column of the table
df1= df.drop(df.columns[[-1]],axis = 1)
```

```python
#print out this dataframe
df1
```

```python
#rename column's titles
df1.columns= ["SEX", "AGE", "COUNTRY", "TIME_PERIOD", "PERCENTAGE"]
```

```python
#print out this dataframe
df1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create plot for one country 
ax = df1[(df1.COUNTRY=='EL')].plot(kind= "line", linestyle= "--", x='TIME_PERIOD',y='PERCENTAGE',color='blue')
df1[(df1.COUNTRY=='EL')].plot(kind= "line", linestyle= "--", x='TIME_PERIOD',y='PERCENTAGE',color='blue',ax=ax)
plt.title("suicides in Greece")
plt.xlabel("TIME_PERIOD")
plt.ylabel("PERCENTAGE")
plt.legend(["percentage of ages 15-19"])
plt.ylim(0,10)
```

```python
#create plot for one country
ax = df1[(df1.COUNTRY=='PT')].plot(kind='line', linestyle= "--",x='TIME_PERIOD',y='PERCENTAGE',color='green')

df1[(df1.COUNTRY=='PT')].plot(kind='line', linestyle= "--",x='TIME_PERIOD',y='PERCENTAGE',color='green',ax=ax)
plt.title("suicides in Portugal")
plt.xlabel("TIME_PERIOD")
plt.ylabel("PERCENTAGE")
plt.legend(["percentage of ages 15-19"])
plt.ylim(0,10)
```

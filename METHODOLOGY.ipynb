Γ) Για την ανάλυση των δεδομένων μας, αρχικά, εστιάσαμε στην έρευνα μεταξύ των ποσοστών των αυτοκτονιών στο κάθε φύλο και συγκεκριμένα σε ποιο από τα δύο φύλα(άνδρας, γυναίκα) παρατηρούνται υψηλότερα ποσοστά. Η απάντηση δόθηκε βασιζόμενες στην ανάλυση συγκεκριμένων περιπτώσεων, δηλαδή χωρών με παρόμοιο πληθυσμό (Ανάλυση 4 χωρών: Ελλάδα, Πορτογαλία, Σουηδία, Τσεχία).    Ύστερα, άξιο αναφοράς ήταν το υψηλότερο ποσοστό που παρατηρείται κατά μία συγκεκριμένη χρονολογία σε μία συγκεκριμένη χώρα, στα δεδομένα που έχουμε συλλέξει. Προχωρώντας, στην ανάλυση και του δεύτερου dataset μας, σχετικά με τα ποσοστά αυτοκτονιών στις ηλικίες 15-19, παρατηρήσαμε πως το συγκεκριμένο ερευνητικό ερώτημα μας δίνει απάντηση και για το ποσοστό αυτοκτονιών της συγκεκριμένης ηλικιακής κλίμακας, κατά το συγκεκριμένο έτος και την συγκεκριμένη χώρα που εμφανίζεται το υψηλότερο ποσοστό αυτοκτονιών.      Παραμένοντας στο δεύτερο dataset μας, ερευνήσαμε πως κυμαίνονται τα ποσοστά αυτά στις συγκεκριμένες ηλικίες στις 4 χώρες που εξετάσαμε βάσει πληθυσμού και στην περίπτωση των φύλων.

Έτσι, επιλέξαμε 4 χώρες με παρόμοιο πληθυσμό ώστε και η ανάλυση μας να είναι πιο εύκολα συγκρίσιμη.  (Σουηδία- 10,099,265, Πορτογαλία- 10,196,709, Ελλάδα- 10,423,054, Τσεχία- 10,708,981). Στην περίπτωση των ποσοστών αυτοκτονιών στην ηλικιακή κλίμακα 15-19 επιλέξαμε τα ραβδογράμματα για τις απεικονίσεις των δεδομένων μας, όπου θα έδειχναν πως αυξάνονται ή μειώνονται ανά χώρα και έτος.
